```python
from datascience import *
# datascience version number of last run of this notebook
print("Data Science Module Version : ", version.__version__)

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from urllib.request import urlopen
from collections import Counter
import numpy as np

import re 


def read_url(url):
    return re.sub('\\s+', ' ', urlopen(url).read().decode())
```

```python
little_women_url = 'http://www.gutenberg.org/cache/epub/514/pg514.txt'
little_women_text = read_url(little_women_url)
chapters = little_women_text.split('CHAPTER')[1:]
```

```python
print(chapters[0])
```

```python
Table().with_columns([
        "Chapters", chapters
    ])
```

```python
Table().with_columns([
        "Chapters", chapters,
        "Jo", np.char.count(chapters,'Jo'),
        "Meg", np.char.count(chapters,'Meg'),
        "Amy", np.char.count(chapters,'Amy'),
        "Beth", np.char.count(chapters,'Beth'),
    ])
```

```python
Table().with_columns([
        "Jo", np.char.count(chapters,'Jo'),
        "Meg", np.char.count(chapters,'Meg'),
        "Amy", np.char.count(chapters,'Amy'),
        "Beth", np.char.count(chapters,'Beth'),
        "Laurie", np.char.count(chapters,'Laurie'),
    ]).cumsum().plot(overlay=True)
```

```python
Table().with_columns([
        'Character',[len(c) for c in chapters],
        'Periods',np.char.count(chapters,'.'),
    ]).scatter('Periods') 
```
